# Using Linear Regression to Classify Diabetes Data

```python
# Using linear regression to classify diabetes data
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
dataset = pd.read_csv('../dataset/diabetes_norm.csv')
```

```python
# Split dataset into training and test set
X = dataset.iloc[:, 0:len(dataset.columns)-1].values
y = dataset.iloc[:, len(dataset.columns)-1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Build linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predict test set results
y_pred = regressor.predict(X_test)
```

```python
# Calculate mean squared error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print(mse)
```

```python
print(y_test)
# Plot training and test set results
for attr1 in range(len(dataset.columns)-2):
    for attr2 in range(attr1+1, len(dataset.columns)-1):
        plt.figure(figsize=(10, 8))
        # Scatter point by point
        for i in range(len(X_test)):
            plt.scatter(X_test[i][attr1], X_test[i][attr2], color=('blue' if y_test[i] == 0 else 'red'))

        # plot regression line intercept and slope
        plt.plot(X_test[:, attr1], regressor.coef_[attr1]*X_test[:, attr1] + regressor.intercept_, color='blue')

        plt.title('Test set')
        plt.xlabel(dataset.columns[attr1])
        plt.ylabel(dataset.columns[attr2])
        plt.show()
```
