```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
# Load dataset
dataset = pd.read_csv('../dataset/diabetes_norm.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
## Missing Data Checking:
dataset.isnull().sum()
```

```python
all_features = dataset.keys()
n_features = all_features[1:7]
label = all_features[7]
```

```python
dat = dataset[n_features]
dat
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dat, dataset[label].values, test_size=0.20, random_state=1)
```

```python
# Train the SVC
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
```

```python
# Prediction
predictions = model.predict(X_test)
print(predictions)
```

```python
# Evaluations
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
